# Get all the job offers

```python
import requests
from bs4 import BeautifulSoup

# Step 1: Fetch the page
url = "https://www.artefact.com/careers/explore-our-jobs/"
response = requests.get(url)
html_content = response.text  # the raw HTML

# Step 2: Parse the HTML
soup = BeautifulSoup(html_content, "html.parser")

# Step 3: Extract data
title = soup.title.text
print(title)
```

```python
urls_job = {}
```

```python
# Create a dictionary for the first page 

urls_job[1] = []
```

```python
# Find job offer

div_job_offer = soup.find_all('div', class_ = "ds_secondary_button")

for job_offer in div_job_offer : 

    urls_job[1].append(job_offer.find("a")["href"])


print(f"We found {len(urls_job[1])} job offers at Artefact on the first page !\nHere are the links : {urls_job[1]}.")

```

# Get the information from the offer

```python
# Try to extract content with the first link. 

link_1st_job = urls_job[1][0]

response_job_1st = requests.get(link_1st_job)
html_content = response_job_1st.text  # the raw HTML

# Step 2: Parse the HTML
soup_job_1st = BeautifulSoup(html_content, "html.parser")
```

```python
div_job_description = soup_job_1st.find('div', class_ = "post-content")
```

```python
# Extract all text inside

text_description_1st = ""
if div_job_description:
    text_content = div_job_description.get_text(separator="\n", strip=True)
    text_description_1st += text_content
```

```python
print(text_description_1st)
```
